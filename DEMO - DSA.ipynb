## DEMO: Time Series Analysis and Forecasting

* Problem Statement: An Airline Company want to forecast the number passengers recurrently for the next 3 years for their capacity planning.
* Goal: To train 2 Forecasting Models using the historical AirLine Passenger data and pick the best model

#### Import the needed Python packages

```python
import pandas as pd
import numpy as np
%matplotlib inline

from statsmodels.tsa.holtwinters import ExponentialSmoothing

from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error

import warnings
warnings.filterwarnings('ignore')
```

#### Import the Historical Airline Passengers data

```python
airline = (pd.read_csv("airline_passengers.csv", index_col = "Month", parse_dates= True)
                .dropna()
            )
airline.index.freq = "MS"
airline.head()
```

#### Explore the Airline Passengers data

```python
airline.plot(figsize = (12, 6))
```

### Modelling

#### Train - Test Split

```python
Train_data = airline.iloc[:109] #Training Set
Test_data =  airline.iloc[108:] #Testing Set
```

---

#### Model Training: Holt Winters Model

```python
fitted_model = ExponentialSmoothing(Train_data["Thousands of Passengers"], 
                                   trend = "add",
                                   seasonal = "mul",
                                   seasonal_periods= 12).fit()
```

#### Generate Forecast from Holt-Winters

```python
HW_predictions = fitted_model.forecast(36)
```

#### Plot the TRAIN and TEST data

```python
Train_data["Thousands of Passengers"].plot(legend = True, 
                                           label = "TRAIN DATA",
                                          figsize = (12, 6))
Test_data["Thousands of Passengers"].plot(legend = True, 
                                           label = "TEST DATA",
                                          figsize = (12, 6))
```

#### Holt-Winters: Performance on predicting the TEST SET

```python
Train_data["Thousands of Passengers"].plot(legend = True, 
                                           label = "TRAIN DATA",
                                          figsize = (12, 6))
Test_data["Thousands of Passengers"].plot(legend = True, 
                                           label = "TEST DATA",
                                          figsize = (12, 6))
HW_predictions.plot(legend = True, label = "HOLT-WINTERS Predictions")
```

#### Compare this MAE value with the mean of the Test_data

```python
mean_absolute_error(Test_data, HW_predictions) 
```

---

#### Model Training: SARIMA

```python
SARIMA = auto_arima(airline["Thousands of Passengers"], 
                    seasonal = True, 
                    m = 12, 
                    trace = False)
# SARIMA.summary()
```

```python
model = SARIMAX(airline["Thousands of Passengers"], 
                order = (2,1,2),
               seasonal_order = (0, 0 , 1, 12))
results = model.fit()
# results.summary()
```

#### Generate Predictions from SARIMA Model

```python
start = len(Train_data)
end = len(Train_data) + len(Test_data) - 1
SARIMA_predictions = (results
                .predict(start, 
                              end, 
                              typ = "levels")
                .rename("SARIMA Predictions"))
```

#### SARIMA: Performance on predicting the TEST DATA

```python
Train_data["Thousands of Passengers"].plot(
            legend = True,
            figsize = (12, 6),
           label = "TRAIN DATA"
)
Test_data["Thousands of Passengers"].plot(
            legend = True,
            figsize = (12, 6),
            label = "TEST DATA"
)
SARIMA_predictions.plot(
            legend = True,
            figsize = (12, 6),
            label = "SARIMA Predictions"
) 
```

#### Mean Absolute Error: SARIMA

```python
mean_absolute_error(Test_data["Thousands of Passengers"], SARIMA_predictions)
# mean_absolute_error(Test_data, test_predictions) 
```
